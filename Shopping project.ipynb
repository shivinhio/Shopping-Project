```python
def outlier_treatment(col):
    ''' calculating outlier indices and replacing them with NA  '''
    #Extract quartiles
    q75, q25 = np.percentile(train_final[col], [75 ,25])
    print(q75,q25)
    #Calculate IQR
    iqr = q75 - q25
    #Calculate inner and outer fence
    minimum = q25 - (iqr*1.5)
    maximum = q75 + (iqr*1.5)
    print(minimum,maximum)
    #Replace with NA
    train_final.loc[train_final[col] < minimum,col] = np.nan
    train_final.loc[train_final[col] > maximum,col] = np.nan
```

```python
outlier_treatment('Purchase')
```

```python
train_final.isnull().sum()
```

```python
train_final['Purchase']= train_final['Purchase'].fillna(train_final['Purchase'].mean())
```

```python
from sklearn.model_selection import train_test_split
X=train_final.drop(['Purchase'],axis=1)
Y=train_final['Purchase']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.50,random_state=101)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtree=DecisionTreeRegressor()
dtree.fit(X_train,Y_train)
```

```python
pred_DT=dtree.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor(n_estimators=30,random_state=101)
rfr.fit(X_train,Y_train)
pred_rfr=rfr.predict(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
pred_lr=lr.predict(X_test)
```

```python
'Coefficients:\n',lr.coef_
```

```python
plt.scatter(Y_test,pred_lr)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
feat_importances=pd.Series(rfr.feature_importances_,index=X.columns)
feat_importances.nlargest(10).plot(kind='bar')
plt.ylabel('Importance')
plt.xlabel('Variable')
plt.title('Variable Importances')
```

```python
def MAPE(Y_test,pred_DT):
    mape=np.mean(np.abs((Y_test-pred_DT)/Y_test)*100)
    return mape
```

```python
MAPE(Y_test,pred_rfr)
```

```python
MAPE(Y_test,pred_DT)
```
